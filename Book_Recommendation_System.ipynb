## **Data  Visualization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 **Histogram to visualize the distribution of ratings**

```python
d = r'/content/drive/MyDrive/Ratings_modified.csv'
df = pd.read_csv(d)
# Display the first few rows of the dataset
print(df.head())
# Create a histogram to visualize the distribution of ratings
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Book-Rating', bins=5, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Book-Rating')
plt.ylabel('Frequency')
plt.show()
```

**visualize the number of ratings per book**

```python
# Create a countplot to visualize the number of ratings per book
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='ISBN', order=df['ISBN'].value_counts().index[:10])
plt.title('Top 10 Most Rated Books')
plt.xlabel('ISBN')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=90)
plt.show()
```

**visualize the number of ratings per user**

```python
# Create a countplot to visualize the number of ratings per user
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='User-ID', order=df['User-ID'].value_counts().index[:10])
plt.title('Top 10 Most Active Users')
plt.xlabel('User-ID')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=90)
plt.show()
```

**Distribution of Books by Publisher**

```python
d=r'/content/drive/MyDrive/Books_modified.csv'
data=pd.read_csv(d)
df = pd.DataFrame(data)
year_counts = df['Year-Of-Publication'].value_counts().sort_index()
print(year_counts)
plt.figure(figsize=(10, 6))
plt.pie(year_counts.values, labels=year_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Books by Publisher')
plt.tight_layout()
plt.show()
```

**Top 20 Authors by Number of Books Published**

```python
author_counts = df.groupby('Book-Author')['ISBN'].count()
print(author_counts)

author_counts = author_counts.sort_values(ascending=False)
plt.figure(figsize=(12, 6))
author_counts[:20].plot(kind='bar')
plt.xlabel('Author')
plt.ylabel('Number of Books Published')
plt.title('Top 20 Authors by Number of Books Published')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
ac=df['Book-Author'].nunique()
print(ac)
```

## **Merging Datasets**

```python

r=r"/content/drive/MyDrive/Ratings_modified.csv"
ratings=pd.read_csv(r)
b=r"/content/drive/MyDrive/Books_modified.csv"
books=pd.read_csv(b)
u=r"/content/drive/MyDrive/Users_modified.csv"
user=pd.read_csv(u)
rating_with_bookname=ratings.merge(books,on='ISBN')
df=pd.DataFrame(rating_with_bookname)
print(df)
num_ratings_df=rating_with_bookname.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
print(num_ratings_df)
```

```python
avg_rating_df=rating_with_bookname.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
print(avg_rating_df)
```

```python
popular_df=num_ratings_df.merge(avg_rating_df,on='Book-Title')
df1=pd.DataFrame(popular_df)
print(df1)
```

```python
popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df.merge(books,on='Book-Title')
```

```python
x=rating_with_bookname.groupby('User-ID').count()['Book-Rating']>200
num_reading_users=x[x].index
filtered_rating=rating_with_bookname[rating_with_bookname['User-ID'].isin(num_reading_users)]
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

**Top 50 Highly Rated Books with More Than 250 Ratings**

```python
top_rated_books = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(20)

plt.figure(figsize=(12, 6))
plt.barh(top_rated_books['Book-Title'], top_rated_books['avg_ratings'], color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Book Title')
plt.title('Top 50 Highly Rated Books with More Than 250 Ratings')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest-rated book at the top
plt.tight_layout()
plt.show()
```

**Word Cloud of Highly Rated Book Titles**

```python

from wordcloud import WordCloud

# Create a string of highly rated book titles
titles_text = ' '.join(final_ratings['Book-Title'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(titles_text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Highly Rated Book Titles')
plt.tight_layout()
plt.show()
```
